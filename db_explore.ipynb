## First, the list of datasets should be created.

```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("db/archive")

files = []
for f in DATA_DIR.rglob("*"):
    if f.is_file():
        files.append({
            "path": str(f),
            "name": f.name,
            "suffix": f.suffix,
            "size_mb": round(f.stat().st_size / 1024**2, 2)
        })

inventory = pd.DataFrame(files)
inventory.sort_values("path").head(50)
```

## Now, the list of only .csv files

```python
csv_files = inventory[inventory["suffix"].str.lower() == ".csv"]
csv_files[["path", "name", "size_mb"]]
```

## Files about policing incident records

```python
keywords = ["uof", "force", "incident", "police", "stop", "prepped"]

candidate_files = csv_files[
    csv_files["name"].str.lower().apply(
        lambda x: any(k in x for k in keywords)
    )
]

candidate_files[["path", "name", "size_mb"]]
```
